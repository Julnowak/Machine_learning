# Zadanie 11

Wyświetl histogramy dla poszczególnych parametrów z uwzglednieniem przynależności do danej klasy.

```python
xlabels = ['sepal length [cm]', 'sepal width [cm]', 'petal length [cm]', 'petal width [cm]']

fig, ax = plt.subplots(2,2, figsize=(20,20))

for i in range(3):
    sl = iris['sepal length (cm)'][iris['target']==float(i)]
    ax[0][0].hist(sl)

    sw = iris['sepal width (cm)'][iris['target']==float(i)]
    ax[0][1].hist(sw)

    pl = iris['petal length (cm)'][iris['target']==float(i)]
    ax[1][0].hist(pl)

    pw = iris['petal width (cm)'][iris['target']==float(i)]
    ax[1][1].hist(pw)

counter = 0
for x in range(2):
    for y in range(2):
        ax[x][y].grid(axis='y')
        ax[x][y].legend(['Class 0.0','Class 1.0','Class 2.0'])
        ax[x][y].set_xlabel(xlabels[counter].capitalize(), fontsize=18)
        ax[x][y].set_ylabel('Count', fontsize=18)
        counter += 1 

        
fig.suptitle('Histograms of parameters for each class').set_fontsize(30)
fig.subplots_adjust(top=0.92)
plt.show()
```
